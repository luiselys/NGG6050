```python
import matplotlib.pyplot as plt
import pandas as pd


# Ct values for gene 1
control_gene1 = [1, 2, 3]
disease_gene1 = [10, 13, 15]

# Ct values for gene 2
control_gene2 = [3, 4, 5]
disease_gene2 = [20, 18, 19]

#genes = pd.read_csv()

# Create a box plot for gene 1
plt.subplot(1, 2, 1)  # Create a subplot for gene 1
plt.boxplot([control_gene1, disease_gene1], labels=['Control', 'Disease'])
plt.title('Gene 1 Ct Values')
plt.ylabel('Ct Values')

# Create a box plot for gene 2
plt.subplot(1, 2, 2)  # Create a subplot for gene 2
plt.boxplot([control_gene2, disease_gene2], labels=['Control', 'Disease'])
plt.title('Gene 2 Ct Values')
plt.ylabel('Ct Values')

plt.tight_layout()  # Ensure proper spacing between subplots
plt.show()
```
