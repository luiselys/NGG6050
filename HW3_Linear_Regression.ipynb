# Definitions

Linear regression is a common way to measure associations between a dependent, measured random variable ($Y$) and a set of independent predictors ($X$).

**Simple linear regression** uses just one predictor:

$\quad y_i=\beta_0+\beta_1x_1+\epsilon_i$

where $\beta_0$ and $\beta_1$ are the "beta weights" describing the y-intercept and slope, respectively, of the line describing the relationship between $X$ and $Y$, and $\epsilon_i$ tells us that the relationship is noisy (more specifically, that the measured linear relationship is corrupted by additive Gaussian noise).

**Multiple linear regression** uses multiple predictors:

$\quad y_i=\beta_0+\beta_1x_1+\beta_2x_2\:+ ... +\:\epsilon_i$

**General linear models**, which we will not consider further here, are even more general and allow the value of $Y$ to be a vector and not just a scalar on each trial.

We typically use linear regression to do any or all of the following:

1\. Infer if a measured variable $Y$ is associated with another variable (predictor) of set of variables (predictors) $X$. This is a hypothesis test. For example, to test if there is a positive relationship between two variables: as you are given more ice cream to eat ($X$ variable), does your body weight ($Y$ variable) change (increase or decrease)? A *p*-value can be assigned to this association; more below on this.

2\. What is the strength of the association? How tightly are these two variables associated? A positive beta weight implies a positive relationship (both go up/down at the same time), whereas a negative value implies a negative relationship (they go in opposite directions).

It is critical to remember that unlike a correlation coefficient, whose magnitude always varies between -1 and 1 independently of the magnitude of the quantities being correlated, in linear regression the magnitude of any given beta weight depends critically on the units and magnitudes of the quantities used in the regression. Therefore, comparing beta weights from two different regressions (or even from different terms in the same multiple linear regression equation, if the predictors are not independent!) is not always straightforward, as discussed [here](http://www.glmj.org/archives/articles/Ziglari_v43n2.pdf).

One way around this problem is to use standardized linear regression, in which the same linear model is used but applied to standardized variables. That is, each of the independent and dependent variables are converted in units of z-score: subtract the sample mean and divide by the sample standard deviation.

3\. The equation that describes the relationship can be used to predict unknown values. That is, once the values of the betas in the above equations are known, they can be used with new values of the independent variables to predict the value of the dependent variable under those conditions.

There are some assumptions in order to use linear regression. For example, for any value of $X$ there exists a normal distribution of $Y$ values, and the relationship between $X$ and $Y$ is linear. Make sure that you always look at your data first (e.g., in scatterplots) to get a first-pass impression if these assumptions are reasonable.

# Calculating the slope of the regression line and its intercept

The **slope** $b=\frac{\sum{X_iY_i-\frac{\sum{X_i}\sum{Y_i}}{n}}}{\sum{X^2_i-\frac{(\sum{X_i})^2}{n}}}$, where *i* is the individual values of $X$ an $Y$ and *n* is the number of observations.  This calculation is the best fit or best estimate of the slope, given your measurements. It is best in the sense that it tries to minimize the distance (in a least squares sense) between $Y_{pred}$ and all $Y_i$.

The **intercept** $a=Y_{avg} - bX_{avg}$, where $Y_{avg}$ and $X_{avg}$ are the average values of $X$ and $Y$.

# Null hypotheses for linear regression


For the slope, this test can be conceptualized in two ways:

1\. $H_0: b=0$. Here an [analysis of variance](https://colab.research.google.com/drive/1wPjM-On_ktPzcV4xv5iP5nW8t0IPBWz1?usp=sharing) and an *F*-statistic ($df=n-2$) is used to support or refute the null hypothesis.

2\. (More general) $H_0:b=b_0$, where $b_0$ can be any value. Here, a 2-tailed [*t*-test](https://colab.research.google.com/drive/1M7xjaMwJUEyULPHfXc3tWG6-WVjCl-uQ?usp=sharing) can be used to test the null hypothesis. 1-tailed *t*-tests can also be used to test null hypotheses such as $H_0:b≤0$ or $H_0:b≥0$.



# Confidence intervals in linear regression

Confidence intervals in linear regression are based on the standard errors of the estimates of the regression parameters and can be computed using the [*t* distribution](https://colab.research.google.com/drive/1Q_Du5NK71Rc1qu-byh8dy8Fs39uvR_4n?usp=sharing):

For the slope, the $100(1-\alpha)$ confidence interval is defined as:

$\quad b\pm t_{\alpha/2,n-2}\times\sqrt{\frac{MSE}{\sum(x_i-\bar{x})^2}}$, where MSE is the mean squared error [defined in the ANOVA tutorial](https://colab.research.google.com/drive/1wPjM-On_ktPzcV4xv5iP5nW8t0IPBWz1?usp=sharing).

For the intercept, the $100(1-\alpha)$ confidence interval is defined as:

$\quad a\pm t_{\alpha/2,n-2}\times\sqrt{\frac{MSE}{n}}$.

## Example 1: London Taxi Drivers

Perhaps one of the more famous early studies using MRI and fMRI to relate behavior to underlying neuroanatomy are the studies that looked at relationships between experience as a taxi-cab drivers in London and hippocampus size. In [the first study](https://www.pnas.org/content/97/8/4398), they found a correlation between hippocampal size (in the posterior region]) and time as a taxi-cab driver. In London, taxi drivers have to take a test showing their knowledge of the streets of London. Here is one of their key findings:

![](https://drive.google.com/uc?export=view&id=1ozXY-A4yvF8NNgY6jwD8ZuiIjsjQwYjZ)

**Legend**: The volume of gray matter in the right hippocampus was found to correlate significantly with the amount of time spent learning to be and practicing as a licensed London taxi driver, positively in the right posterior hippocampus.

Without getting into details of the science itself, do you think a liner regression is a proper way to analyze this data set? Are the data linear? Why or why not? If not, how does that weaken any interpretation of the data?

In a [follow-up study](https://onlinelibrary.wiley.com/doi/epdf/10.1002/hipo.20233), they compared taxi cab drivers with bus drivers and found a correlation between years of driving experience and hippocampal size for the taxi cab drivers but not for the bus drivers. They argued that the basis for these results is the observation that bus drivers drive the same route everyday, whereas taxi cab drivers have to know the entire city and be flexible in how they navigate the city. Can one make any sort of inference about casualty from this finding? What are the weaknesses and strengths of such a study and set of findings?



Analyzing the relationship between the volume of gray matter in the right hippocampus and the amount of time spent learning to be and practicing as a licensed London taxi driver using linear regression may not be appropriate due to the potential non-linearity of the relationship. Here are some reasons why the data might not be linear:

Threshold Effects: The relationship between the time spent as a taxi driver and hippocampal volume may not be linear. It's possible that there are threshold effects, meaning that there might be a minimum level of experience required before any significant changes in hippocampal volume occur. After reaching this threshold, further experience may not lead to proportional increases in hippocampal volume.

Saturation Effects: It's also possible that there is a saturation point, where additional experience beyond a certain point does not result in further changes in hippocampal volume. In other words, the relationship might be nonlinear and reach a plateau.

Diminishing Returns: Initially, as taxi drivers gain experience and navigate London's streets, the hippocampus might undergo significant structural changes. However, over time, the rate of change in hippocampal volume may decrease, leading to diminishing returns. This would also suggest a nonlinear relationship.

If the relationship between hippocampal volume and experience as a taxi driver is indeed nonlinear, attempting to fit a linear regression model to the data could lead to misinterpretations and biased estimates. Linear regression assumes a linear relationship between the independent and dependent variables, and if this assumption is violated, the model may not accurately represent the true nature of the relationship.

To address potential nonlinearity, more sophisticated statistical techniques may be needed, such as polynomial regression, spline regression, or nonparametric regression methods. These approaches can capture more complex patterns in the data and provide a better understanding of how hippocampal volume changes with increasing experience as a taxi driver.

In summary, while linear regression can be a useful tool for analyzing data when there is a linear relationship between variables, it's important to consider the nature of the data and the possibility of nonlinear relationships. In cases like the one described, where the relationship may not be linear, using alternative modeling approaches can lead to more accurate and meaningful interpretations.

The findings of the follow-up study, which compared taxi cab drivers and bus drivers in terms of the correlation between years of driving experience and hippocampal size, provide valuable insights into the potential effects of specific types of driving experience on brain structure. However, it is essential to carefully consider the limitations and strengths of this study and avoid making strong causal inferences. Here's an evaluation of the study's weaknesses and strengths:

Weaknesses:

Correlation vs. Causation: The study identifies a correlation between years of driving experience and hippocampal size for taxi cab drivers but not for bus drivers. While this correlation is interesting, it does not establish causation. The study cannot definitively conclude that driving experience caused changes in hippocampal size. There could be other factors at play, such as pre-existing differences in brain structure or lifestyle factors.

Selection Bias: The study's sample may not be representative of all taxi cab and bus drivers. There could be selection biases that influence the findings. For example, individuals who choose to become taxi cab drivers might have different cognitive or lifestyle characteristics than those who choose to become bus drivers.

Lack of Randomization: The study likely lacks randomization, meaning that participants self-select their profession (taxi cab or bus driver). Random assignment to groups is a powerful technique for minimizing confounding variables, which may not have been adequately addressed in this observational study.

Third-Variable Problem: There may be third variables (confounding factors) that explain the observed differences. For example, it's possible that the taxi cab drivers and bus drivers differ in other ways, such as stress levels, lifestyle, or cognitive abilities, which could affect both their choice of profession and their brain structure.

Strengths:

Naturalistic Observation: The study benefits from a naturalistic setting where participants are in their real-world work environments. This allows for the examination of real-world experiences and their potential impact on brain structure.

Hypothesis-Driven: The study was conducted based on a hypothesis derived from the earlier findings with London taxi drivers. It was designed to test whether similar associations between driving experience and brain structure existed in a different group of drivers.

Control Group: The inclusion of bus drivers as a control group is a strength because it allows for a comparison to be made between two groups with different driving experiences, helping to isolate the potential effects of the type of driving experience.

Replication: The findings, if consistent with earlier research on London taxi drivers, add to the body of evidence suggesting a link between navigation demands and brain structure. Replicating findings in different populations strengthens the overall argument.

In summary, while the study provides intriguing insights into the potential relationship between driving experience and hippocampal size, it is essential to exercise caution when inferring causality from correlation in observational studies. The findings suggest that there may be differences between taxi cab and bus drivers, but they do not establish a clear causal relationship. Additional research, ideally through controlled experiments or longitudinal studies, would be needed to explore causation and better address potential confounding factors.

##### 2\. Calculate and plot the regression line.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
age = np.array([3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17])
wing_length = np.array([1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5.0])

# Calculate the regression line
slope, intercept = np.polyfit(age, wing_length, 1)

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(age, wing_length, color='blue', marker='o', label='Age vs. Wing Length')

# Plot the regression line
regression_line = slope * age + intercept
plt.plot(age, regression_line, color='red', label='Regression Line')

# Add labels and a title
plt.xlabel('Age')
plt.ylabel('Wing Length')
plt.title('Relationship Between Age and Wing Length with Regression Line')

# Show a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

# Print the regression equation
print(f"Regression Equation: Wing Length = {slope:.2f} * Age + {intercept:.2f}")
```

##### 3\. Can you reject $H_0:b=0$?

```python
import numpy as np
from scipy.stats import linregress

# Data
age = np.array([3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17])
wing_length = np.array([1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5.0])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(age, wing_length)

# Set the significance level
alpha = 0.05

# Print the p-value
print(f"P-value: {p_value:.4f}")

# Check if the p-value is less than the significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant relationship (b ≠ 0)")
else:
    print("Fail to reject the null hypothesis: There is no significant relationship (b = 0)")
```

##### 4\. Calculate and plot the confidence intervals on the slope of the regression.

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Data
age = np.array([3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17])
wing_length = np.array([1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5.0])

# Add a constant term (intercept) to the independent variable
age_with_intercept = sm.add_constant(age)

# Perform linear regression
model = sm.OLS(wing_length, age_with_intercept).fit()

# Get the confidence intervals for the slope
confidence_intervals = model.conf_int(alpha=0.05)  # 95% confidence interval

# Extract the slope and intercept
slope = model.params[1]
intercept = model.params[0]

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(age, wing_length, color='blue', marker='o', label='Age vs. Wing Length')

# Plot the regression line
regression_line = slope * age + intercept
plt.plot(age, regression_line, color='red', label='Regression Line')

# Plot confidence intervals
lower_ci = (confidence_intervals[1, 0] * age) + confidence_intervals[0, 0]
upper_ci = (confidence_intervals[1, 1] * age) + confidence_intervals[0, 1]
plt.fill_between(age, lower_ci, upper_ci, color='gray', alpha=0.5, label='Confidence Intervals (95%)')

# Add labels and a title
plt.xlabel('Age')
plt.ylabel('Wing Length')
plt.title('Relationship Between Age and Wing Length with Confidence Intervals')

# Show a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

# Print the confidence intervals
print(f"95% Confidence Intervals for Slope:")
print(f"Lower Bound: {confidence_intervals[1, 0]:.4f}")
print(f"Upper Bound: {confidence_intervals[1, 1]:.4f}")
```

##### 4\. Calculate $r^2$ (the coefficient of determination)

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Data
age = np.array([3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17])
wing_length = np.array([1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5.0])

# Add a constant term (intercept) to the independent variable
age_with_intercept = sm.add_constant(age)

# Perform linear regression
model = sm.OLS(wing_length, age_with_intercept).fit()

# Get the R-squared value
r_squared = model.rsquared

# Extract the slope and intercept
slope = model.params[1]
intercept = model.params[0]

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(age, wing_length, color='blue', marker='o', label='Age vs. Wing Length')

# Plot the regression line
regression_line = slope * age + intercept
plt.plot(age, regression_line, color='red', label='Regression Line')

# Plot confidence intervals
lower_ci = (model.conf_int(alpha=0.05)[1, 0] * age) + model.conf_int(alpha=0.05)[0, 0]
upper_ci = (model.conf_int(alpha=0.05)[1, 1] * age) + model.conf_int(alpha=0.05)[0, 1]
plt.fill_between(age, lower_ci, upper_ci, color='gray', alpha=0.5, label='Confidence Intervals (95%)')

# Add labels and a title
plt.xlabel('Age')
plt.ylabel('Wing Length')
plt.title('Relationship Between Age and Wing Length with Confidence Intervals')

# Show a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

# Print the R-squared value
print(f"R-squared (Coefficient of Determination): {r_squared:.4f}")
```

##### 6\. Calculate Pearson's *r*.

```python
import numpy as np

# Data
age = np.array([3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17])
wing_length = np.array([1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5.0])

# Calculate Pearson's correlation coefficient
pearson_r = np.corrcoef(age, wing_length)[0, 1]

# Print the calculated Pearson's r
print(f"Pearson's Correlation Coefficient (r): {pearson_r:.4f}")
```

##### 7\. Add some noise to the data and see how the regression changes.

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Data
age = np.array([3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17])
wing_length = np.array([1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5.0])

# Add noise to the wing length data
np.random.seed(0)  # Set a random seed for reproducibility
noise = np.random.normal(0, 0.5, len(wing_length))
noisy_wing_length = wing_length + noise

# Add a constant term (intercept) to the independent variable
age_with_intercept = sm.add_constant(age)

# Perform linear regression with noisy data
noisy_model = sm.OLS(noisy_wing_length, age_with_intercept).fit()

# Extract the slope and intercept for noisy data
noisy_slope = noisy_model.params[1]
noisy_intercept = noisy_model.params[0]

# Calculate the regression line for noisy data
noisy_regression_line = noisy_slope * age + noisy_intercept

# Calculate R-squared for noisy data
noisy_r_squared = noisy_model.rsquared

# Calculate Pearson's correlation coefficient (r) for noisy data
noisy_pearson_r = np.corrcoef(age, noisy_wing_length)[0, 1]

# Perform linear regression with original data
original_model = sm.OLS(wing_length, age_with_intercept).fit()
original_slope = original_model.params[1]
original_intercept = original_model.params[0]
original_regression_line = original_slope * age + original_intercept

# Calculate R-squared for original data
original_r_squared = original_model.rsquared

# Calculate Pearson's correlation coefficient (r) for original data
original_pearson_r = np.corrcoef(age, wing_length)[0, 1]

# Create a scatter plot for noisy data
plt.figure(figsize=(12, 6))

# Original data points
plt.scatter(age, wing_length, color='blue', marker='o', label='Original Data')

# Noisy data points
plt.scatter(age, noisy_wing_length, color='green', marker='x', label='Noisy Data')

# Plot the regression line for noisy data
plt.plot(age, noisy_regression_line, color='red', linestyle='--', label='Noisy Regression Line')

# Plot the regression line for original data
plt.plot(age, original_regression_line, color='purple', linestyle='-', label='Original Regression Line')

# Add labels and a title
plt.xlabel('Age')
plt.ylabel('Wing Length')
plt.title('Comparison of Regression Lines with Noise')

# Show a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

# Print R-squared and Pearson's r for both noisy and original data
print(f"Noisy Data - R-squared: {noisy_r_squared:.4f}, Pearson's r: {noisy_pearson_r:.4f}")
print(f"Original Data - R-squared: {original_r_squared:.4f}, Pearson's r: {original_pearson_r:.4f}")
```
