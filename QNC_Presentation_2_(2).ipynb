```python
import pandas as pd
import numpy as np
from statsmodels.stats.power import TTestIndPower
import matplotlib.pyplot as plt
from google.colab import files

# Upload an Excel file to Google Colab
uploaded = files.upload()

# Specify the file name after uploading
file_name = 'NRTN.xlsx'  # Replace with your file name

# Read the uploaded Excel file into a Pandas DataFrame
data_frame = pd.read_excel(file_name)

# Convert DataFrame to NumPy array
ct_values = data_frame.to_numpy()

# Indexes for the control group and other groups
control_group_index = 0  # Assuming gene1_control is the first column
other_groups_indexes = [1, 2, 3]  # Assuming gene1_disease, gene1_controlwithtreatment, gene1_diseasewithtreatment follow

# Initialize TTestIndPower for power analysis
power_analysis = TTestIndPower()

effect_sizes = []
powers = []

# Labels for annotating the plot
group_labels = ['Gene1 Disease', 'Gene1 Control with Treatment', 'Gene1 Disease with Treatment']

# Colors for each group
colors = ['blue', 'orange', 'green']

# Perform power analysis for each comparison
for idx, other_index in enumerate(other_groups_indexes):
    # Select the control group and other groups from the NumPy array
    control_group = ct_values[:, control_group_index]
    other_group = ct_values[:, other_index]

    # Calculate effect size based on the mean difference between groups
    effect_size = abs(np.mean(control_group) - np.mean(other_group)) / np.sqrt(
        (np.var(control_group) + np.var(other_group)) / 2
    )
    effect_sizes.append(effect_size)

    # Perform power analysis for a two-sample t-test to determine power with known effect size
    power = power_analysis.solve_power(
        effect_size=effect_size,
        nobs1=len(control_group),
        alpha=0.05,
        ratio=1.0,
        alternative='two-sided'
    )
    powers.append(power)

    print(f"Effect size to detect difference between gene1_control and group {other_index}: {effect_size:.4f}")
    print(f"Power with this effect size: {power:.4f}")
#power of .8, what sample size? have estimate of effect size. effect size on x, sample on y
# Plotting power vs. effect size with annotations and colors
plt.figure(figsize=(8, 6))
for i in range(len(effect_sizes)):
    plt.scatter(effect_sizes[i], powers[i], marker='o', color=colors[i], label=group_labels[i])

plt.title('Power Analysis: Power vs. Effect Size')
plt.xlabel('Effect Size')
plt.ylabel('Power')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
import statsmodels.stats.power as smp
from google.colab import files
import warnings

def analyze_and_plot(file_name, alpha=0.05, desired_power=0.8):
    # Read the uploaded Excel file into a Pandas DataFrame
    data_frame = pd.read_excel(file_name)

    # Convert DataFrame to NumPy array
    ct_values = data_frame.to_numpy()

    # Indexes for the control group and other groups
    control_group_index = 0  # Assuming gene1_control is the first column
    other_groups_indexes = [1, 2, 3]  # Assuming glaucoma, controlwithtreatment, glaucomawithtreatment follow

    # Extracting columns for each group and genes of interest with fold change values
    control_fold_change = ct_values[:, control_group_index]
    other_groups_fold_change = ct_values[:, other_groups_indexes]

    # Perform one-way ANOVA
    _, p_value_anova = f_oneway(control_fold_change, *other_groups_fold_change.T)

    # Check the p-value for significance
    if p_value_anova < alpha:
        print(f"The overall ANOVA for {file_name} is significant, so we can proceed with power analysis.")

        # Calculate effect size
        effect_size = np.sqrt(np.sum((np.mean(group_fold_change) - np.mean(control_fold_change)) ** 2
                                     for group_fold_change in other_groups_fold_change.T) / len(other_groups_indexes))

        # Select desired power, alpha, and perform power analysis
        analysis = smp.FTestAnovaPower()

        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            try:
                sample_size = analysis.solve_power(effect_size=effect_size, k_groups=len(other_groups_indexes),
                                                   nobs=None, alpha=alpha, power=desired_power)
                sample_size_rounded = np.round(sample_size)  # Round the sample size
                print(f"To achieve a power of {desired_power} with an effect size of {effect_size:.4f} and alpha of {alpha},"
                      f" you'll need a sample size of approximately {sample_size_rounded}.\n")
                return effect_size, sample_size_rounded
            except:
                print(f"Convergence warning for {file_name}. Skipping power analysis for this dataset.\n")
                return None, None
    else:
        print(f"The overall ANOVA for {file_name} is not significant, so power analysis is not performed.\n")
        return None, None

# Upload multiple Excel files
uploaded_files = files.upload()

# List to store results for each dataset
effect_sizes = []
sample_sizes = []

# Analyze and plot each dataset
for file_name in uploaded_files.keys():
    effect_size, sample_size = analyze_and_plot(file_name)
    if effect_size is not None and sample_size is not None:
        effect_sizes.append(effect_size)
        sample_sizes.append(sample_size)

# Scatter plot of effect sizes and sample sizes for each dataset
plt.scatter(effect_sizes, sample_sizes, color='blue', label='Effect Size vs Sample Size')
plt.xlabel('Effect Size')
plt.ylabel('Sample Size')
plt.title('Effect Size vs Sample Size for Multiple Datasets')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import probplot
from google.colab import files

# Upload an Excel file to Google Colab
uploaded = files.upload()

# Specify the file name after uploading
file_name = 'NRTN.xlsx'  # Replace with your file name

# Read the uploaded Excel file into a Pandas DataFrame
data_frame = pd.read_excel(file_name)

# Convert DataFrame to NumPy array
ct_values = data_frame.to_numpy()

# Indexes for the control group and other groups
control_group_index = 0  # Assuming gene1_control is the first column
other_groups_indexes = [1, 2, 3]  # Assuming glaucoma, controlwithtreatment, glaucomawithtreatment follow

# Extracting columns for each group and genes of interest with fold change values
control_fold_change = ct_values[:, control_group_index]
other_groups_fold_change = ct_values[:, other_groups_indexes]

# Mapping group indexes to their corresponding names
group_names = {
    1: 'Glaucoma',
    2: 'Control with Treatment',
    3: 'Glaucoma with Treatment'
}

# QQ plot for each comparison
fig, ax = plt.subplots(1, len(other_groups_indexes), figsize=(5 * len(other_groups_indexes), 5))

for i, group_index in enumerate(other_groups_indexes):
    probplot(other_groups_fold_change[:, i], dist="norm", plot=ax[i])
    ax[i].set_title(f'Control vs {group_names[group_index]}')

plt.tight_layout()
plt.show()
```

# Shapiro-Wilk test

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro
from google.colab import files

# Upload an Excel file to Google Colab
uploaded = files.upload()

# Specify the file name after uploading
file_name = 'NRTN.xlsx'  # Replace with your file name

# Read the uploaded Excel file into a Pandas DataFrame
data_frame = pd.read_excel(file_name)

# Convert DataFrame to NumPy array
ct_values = data_frame.to_numpy()

# Indexes for the control group and other groups
control_group_index = 0  # Assuming gene1_control is the first column
other_groups_indexes = [1, 2, 3]  # Assuming glaucoma, controlwithtreatment, glaucomawithtreatment follow

# Extracting columns for each group and genes of interest with fold change values
control_fold_change = ct_values[:, control_group_index]
other_groups_fold_change = ct_values[:, other_groups_indexes]

# Mapping group indexes to their corresponding names
group_names = {
    1: 'Glaucoma',
    2: 'Control with Treatment',
    3: 'Glaucoma with Treatment'
}

# Perform Shapiro-Wilk test for each group
shapiro_results = {}
for i, group_index in enumerate(other_groups_indexes):
    _, p_value = shapiro(other_groups_fold_change[:, i])
    shapiro_results[group_names[group_index]] = p_value

# Perform Shapiro-Wilk test for each group and interpret results
shapiro_results = {}
for i, group_index in enumerate(other_groups_indexes):
    _, p_value = shapiro(other_groups_fold_change[:, i])
    shapiro_results[group_names[group_index]] = p_value

    # Interpretation of Shapiro-Wilk test
    if p_value > 0.05:
        interpretation = "Normally distributed"
    else:
        interpretation = "Not normally distributed"

    print(f"{group_names[group_index]} Shapiro-Wilk p-value: {p_value:.4f} - {interpretation}")

# Plot Shapiro-Wilk test results
plt.bar(shapiro_results.keys(), shapiro_results.values())
plt.xlabel('Groups')
plt.ylabel('Shapiro-Wilk p-value')
plt.title('NRTN')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway
from google.colab import files

# Upload an Excel file to Google Colab
uploaded = files.upload()

# Specify the file name after uploading
file_name = 'NRTN.xlsx'  # Replace with your file name

# Read the uploaded Excel file into a Pandas DataFrame
data_frame = pd.read_excel(file_name)

# Convert DataFrame to NumPy array
ct_values = data_frame.to_numpy()

# Indexes for the control group and other groups
control_group_index = 0  # Assuming gene1_control is the first column
other_groups_indexes = [1, 2, 3]  # Assuming glaucoma, controlwithtreatment, glaucomawithtreatment follow

# Extracting columns for each group and genes of interest with fold change values
control_fold_change = ct_values[:, control_group_index]
other_groups_fold_change = ct_values[:, other_groups_indexes]

# Perform ANOVA test for each specific pair
pairwise_tests = {}

for i, group_index in enumerate(other_groups_indexes):
    _, p_value = f_oneway(control_fold_change, other_groups_fold_change[:, i])
    pairwise_tests[f"Control vs {group_index}"] = p_value

# Interpret pairwise ANOVA results
for pair, p_value in pairwise_tests.items():
    if p_value < 0.05:
        print(f"{pair} ANOVA p-value: {p_value:.4f}. There is a significant difference.")
    else:
        print(f"{pair} ANOVA p-value: {p_value:.4f}. There is no significant difference.")
```
